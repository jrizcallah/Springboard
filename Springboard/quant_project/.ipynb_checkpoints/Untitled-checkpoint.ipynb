```python
import pandas as pd
```

```python
data = pd.read_csv('processed_data.csv')
data.drop('Unnamed: 0', axis=1, inplace=True)
data
```

```python
data['close'].max()
```

```python
assets = data['asset'].unique()
for asset in assets:
    temp = data[data['asset'] == asset]
    temp['cci'] = (temp['cci'] - temp['cci'].mean()) / temp['cci'].std()
    data.loc[data['asset'] == asset, 'cci'] = temp['cci']
```
